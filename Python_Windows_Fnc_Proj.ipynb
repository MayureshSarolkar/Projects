```python
import random

# Define product categories and their total sales
categories = {
    "Furniture": 58578.25,
    "Sports": 39743.95,
    "Food": 73104.65
}

# Define subcategories for each category
subcategories = {
    "Furniture": ["chairs", "tables", "wardrobe", "beds", "couch", "antique", "storage"],
    "Sports": ["shoes", "bags", "backpack", "bottles", "towels"],
    "Food": ["chips", "bread", "chocolate", "biscuits", "rice", "wheat", "sugar", "flour"]
}

# Create dictionary to store sale prices for each subcategory
subcategory_prices = {}

# Assign random prices to each subcategory such that their total equals the total sales of their corresponding category
for category, total_sales in categories.items():
    subcats = subcategories[category]
    prices = []
    for subcat in subcats:
        # Generate random price for subcategory
        price = random.uniform(0, total_sales)
        # Add price to list of prices for this category
        prices.append(price)
        # Save price in dictionary
        subcategory_prices[subcat] = price
    # Make sure prices for this category add up to the correct total
    difference = total_sales - sum(prices)
    subcategory_prices[subcats[0]] += difference

# Print sale prices for each subcategory
for subcat, price in subcategory_prices.items():
    print(f"{subcat}: ${price:.2f}")

# Simulate a transaction: sale of a subcategory product
def simulate_transaction(subcat, sale_amount):
    # Update sale prices
    subcategory_prices[subcat] -= sale_amount
    # Update total sales for corresponding category
    category = [c for c, sc in subcategories.items() if subcat in sc][0]
    categories[category] -= sale_amount

# Test simulation of transaction
simulate_transaction("chairs", 100)

# Print updated sale prices for each subcategory
print("\nUpdated sale prices:")
for subcat, price in subcategory_prices.items():
    print(f"{subcat}: ${price:.2f}")

# Print total sales for a given product category
def print_total_sales(category):
    total_sales = categories[category]
    print(f"Total sales for {category}: ${total_sales:.2f}")

# Test printing of total sales
print_total_sales("Furniture")

# Print total sales for a given product subcategory
def print_subcategory_sales(subcat):
    sale_price = subcategory_prices[subcat]
    print(f"Total sales for {subcat}: ${sale_price:.2f}")

# Test printing of subcategory sales
print_subcategory_sales("chairs")
```
